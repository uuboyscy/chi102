```python
with open("stopword.txt", "r", encoding="utf-8") as f:
    stop_word_list = [word.replace("\n", "") for word in f.readlines()]

stop_word_list
```

```python
word_count_dict = {}

for word in word_cut_list:
    if (word in stop_word_list) or (len(word) == 1):
        continue
    if word in word_count_dict:
        word_count_dict[word] += 1
    else:
        word_count_dict[word] = 1
```

```python
word_count_dict
```

```python
word_count_sort_list = []
for k, v in word_count_dict.items():
    word_count_sort_list.append((k, v))

word_count_sort_list = [(k, v) for k, v in word_count_dict.items()]
```
