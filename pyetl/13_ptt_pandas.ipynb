```python
import glob
import pandas as pd
```

```python
"""
[
    [
        "推: 5",
        "噓: 1",
        "分數: 4",
        "作者: gk1329 ()",
        "標題: Re: [新聞] 醫告知婦人「癌症第3期」她拒治療病逝，",
        "時間: Fri Dec 15 20:21:24 2023",
    ],
    [
        "推: 5",
        "噓: 1",
        "分數: 4",
        "作者: gk1329 ()",
        "標題: Re: [新聞] 醫告知婦人「癌症第3期」她拒治療病逝，",
        "時間: Fri Dec 15 20:21:24 2023",
    ], ...
]
"""
```

```python
data = []
for file_path in glob.glob("res_gossiping/*.txt"):
    with open(file_path, "r", encoding="utf-8") as f:
        """
        [
            "推: 5",
            "噓: 1",
            "分數: 4",
            "作者: gk1329 ()",
            "標題: Re: [新聞] 醫告知婦人「癌症第3期」她拒治療病逝，",
            "時間: Fri Dec 15 20:21:24 2023",
        ]
        """
        row_data = f.read().split("---split---")[1].split("\n")
        data.append(row_data[1:-1])

data
```

```python
columns = ["推", "噓", "分數", "作者", "標題", "時間"]

df = pd.DataFrame(data=data, columns=columns)

df
```

```python
def column_transformer(column_value: str) -> int:
    return int(column_value.split(": ")[1])
```

```python
df["推"].apply(column_transformer)
```

```python
df["transformed_push"] = df["推"].apply(column_transformer)
```

```python
df
```

```python
df["transformed_down"] = df["噓"].apply(lambda x: int(x.split(": ")[1]))
```
